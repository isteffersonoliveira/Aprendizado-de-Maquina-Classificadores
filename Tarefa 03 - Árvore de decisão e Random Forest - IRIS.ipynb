Sobre o conjunto de dados
O conjunto de dados Iris foi usado no artigo clássico de 1936 de RA Fisher, The Use of Multiple Measurements in Taxonomic Problems , e também pode ser encontrado no UCI Machine Learning Repository .

Inclui três espécies de íris com 50 amostras cada, bem como algumas propriedades sobre cada flor. Uma espécie de flor é linearmente separável das outras duas, mas as outras duas não são linearmente separáveis ​​uma da outra.

As colunas neste conjunto de dados são:

* Id
* SepalLengthCm
* SepalWidthCm
* PetalLengthCm
* PetalWidthCm
* Species


http://archive.ics.uci.edu/ml/datasets/Iris

## Importação das Bibliotecas

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb


%matplotlib inline
```

## Visualização dos dados

```python
df = pd.read_csv('iris.data', encoding='ISO-8859-1', 
                 names=['sepal length (cm)',
                          'sepal width (cm)',
                          'petal length (cm)',
                          'petal width (cm)',
                          'target'])
```

```python
df
```

```python
df.describe().round(2)
```

```python
sb.pairplot(df, hue='target')
```

```python
X = np.array(df)
X
```

```python
X = np.array(df.drop('target',1))
X
```

```python
y = np.array(df['target'])
y
```

## SEPARANDO DADOS DE TREINO E TESTE

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, y, test_size=0.33, random_state=6)
```

## TREINANDO O MODELO ARVORE DE DECISÃO

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
tree_flores = DecisionTreeClassifier(criterion='entropy', random_state=5)
```

```python
tree_flores.fit(X_train,Y_train)
```

## Ganho de Informação de cada atributo

```python
tree_flores.feature_importances_
```

```python
print(f'sepal length (cm) = {tree_flores.feature_importances_[0].round(2)}')

print(f'sepal width (cm) = {tree_flores.feature_importances_[1].round(2)}')

print(f'petal length (cm) = {tree_flores.feature_importances_[2].round(2)}')

print(f'petal width (cm) = {tree_flores.feature_importances_[3].round(2)}')
```

## Gerando o Gráfico da Arvore de Decisão

```python
tree_flores.classes_
```

```python
from dataclasses import field
from sklearn import tree
previsores = ['petal width (cm)', 'sepal length (cm)', 'sepal width (cm)', 'petal length (cm)']
figura, eixos = plt.subplots(nrows=1, ncols=1, figsize=(15,15))
tree.plot_tree(tree_flores, feature_names=previsores, class_names=tree_flores.classes_, filled=True);
```

## TESTANDO O MODELO

Testando um único valor:

```python
tree_flores.predict([[5.7, 2.8, 4.5, 1.3]])
```

verificando os dados de teste:

```python
TESTE = pd.DataFrame(data=X_test)
TESTE.columns = ['sepal length (cm)',
                          'sepal width (cm)',
                          'petal length (cm)',
                          'petal width (cm)']
TESTE.head()
```

```python
TESTE['real_target'] = Y_test
TESTE.head()
```

```python
TESTE['predict_target'] = tree_flores.predict(X_test)
```

## ITENS QUE NÃO FORAM PREVISTOS CORRETAENTE

```python
TESTE[TESTE['real_target'] != TESTE['predict_target']]
```

## Acurácia do Modelo

```python
from sklearn.metrics import accuracy_score, confusion_matrix, ConfusionMatrixDisplay, classification_report
```

```python
acuracia_iris = accuracy_score(Y_test, tree_flores.predict(X_test))
acuracia_iris
```

## Matrix Confusão

```python
conf_matrix_iris = confusion_matrix(Y_test, tree_flores.predict(X_test))
conf_matrix_iris
```

```python

ConfusionMatrixDisplay(conf_matrix_iris).plot()

plt.show()
```

# Report do Resultado Final

```python
report_iris = classification_report(Y_test, tree_flores.predict(X_test), target_names=["Iris-setosa", "Iris-versicolor", "Iris-virginica"])
```

```python
print(report_iris)
```

## Conclusão Modelo Árvore de Decisão
O modelo previu:

* 100% das espécies Iris-setosa, com 100% de precisão;
* 100% das Iris-versicolor, com 85% de precisão;
* 80% das Iris-virginica, com 100% de precisão.

Acurácia do Modelo: 94%
